## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def X(n):
    return 1/np.sqrt(n) * (np.sum(np.vectorize(lambda x: -1 if x == 0 else x)(np.random.randint(2, size=n))))
def samples(m, n):
    return [X(n) for _ in range(m)]
```

```python
fig, ax = plt.subplots()

x_axis = np.arange(-10, 10, 0.001)

ax.hist(samples(1000, 1000))
ax.plot(x_axis, norm.pdf(x_axis,0,2))
```
