```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
```

```python
def X(n):
    return 1/np.sqrt(n) * (np.sum(np.vectorize(lambda x: -1 if x == 0 else x)(np.random.randint(2, size=n))))
def samples(m, n):
    return np.array([X(n) for _ in range(m)])
```

```python
l = [1, 3, 10, 30, 100, 1000]

fig, ax = plt.subplots(len(l), 1, figsize=(12, 16))
fig.tight_layout()

x_axis = np.arange(-5, 5, 0.001)
for i, x in enumerate(l):
    _ = ax[i].plot(x_axis, norm.pdf(x_axis), c='black')
    _ = ax[i].hist(samples(400, x), density=True)
    ax[i].set_title('$X_{' + str(x) + '}$')
```
