```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./census.csv", sep=";")

fig, ax = plt.subplots()
_ = ax.hist(data['rel_foreign'], bins=100, log=True, label="foreigners")
_ = ax.hist(data['rel_women'], bins=100, label="women")
_ = ax.set_title("distribution of foreigners and women in different communes")
ax.legend()
ax.set_ylabel("number of communes")
ax.set_xlabel("percentage")
```

```python
data['bin'] = pd.qcut(data['Total'],q=10)
```

```python
wdata = []
fdata = []
for b, df in data.groupby('bin', observed=True):
    wdata.append(df['rel_women'])
    fdata.append(df['rel_foreign'])
w_violinparts = plt.violinplot(wdata, bw_method=0.01, widths=0.9, showextrema=False)
f_violinparts = plt.violinplot(fdata, bw_method=0.01, widths=0.9, showextrema=False)
for o in w_violinparts['bodies']:
    o.set_color('red')
    o.set_alpha(0.7)
for o in f_violinparts['bodies']:
    o.set_color('blue')
    o.set_alpha(0.5)
```
