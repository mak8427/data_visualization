```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

# Setup the figure
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

scat5 = ax.scatter(xA[10, :30, 0], xA[10, :30, 1], c='#005555')
scat4 = ax.scatter(xA[8, :30, 0], xA[8, :30, 1], c='#006666')
scat3 = ax.scatter(xA[6, :30, 0], xA[6, :30, 1], c='#007777')
scat2 = ax.scatter(xA[4, :30, 0], xA[4, :30, 1], c='#008888')
scat1 = ax.scatter(xA[2, :30, 0], xA[2, :30, 1], c='#009999')
scat0 = ax.scatter(xA[0, :30, 0], xA[0, :30, 1], c='#00aaaa')

# Update function using the scatter object
def update(frame=0):
    scat4.set_offsets(xA[frame+8, :30, :])
    scat3.set_offsets(xA[frame+6, :30, :])
    scat2.set_offsets(xA[frame+4, :30, :])
    scat1.set_offsets(xA[frame+2, :30, :])
    scat0.set_offsets(xA[frame+0, :30, :])
    fig.canvas.draw_idle()

# Create and display the slider
slider = widgets.IntSlider(min=0, max=40, step=1, description='Frame')
_ = widgets.interact(update, frame=slider)
```
